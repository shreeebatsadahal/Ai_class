```python
dict2={
    "Students":["Student1","Student2","Student3","Student4","Student5","Student1","Student7"],
    "Rollno":['001','002',np.nan,'004','005','001','007'],
    "Gender":["M","F",np.nan,"F","M","M","M"],
    "Marks":[10,20,30,np.nan,50,10,60]

}
```

```python
df2=pd.DataFrame(dict2)
```

```python
df2
```

```python
###TO CHECK THE NULL VALUE
```

```python
df2.isnull().sum()
```

```python
df2.duplicated().sum()
```

```python
###REMOVE THE DUPLICATE VALUE
df2.drop_duplicates()
```

```python
df2.dropna()
```

```python
###MEan and mode imputation

df2
```

```python
gendermode=df2['Gender'].mode()[0]
```

```python
gendermode
```

```python
marksmean=df2['Marks'].mean()
```

```python
marksmean
```

```python
df2['Gender']=df2['Gender'].fillna(gendermode)
df2['Marks']=df2['Marks'].fillna(marksmean)
```
