### Checking missing value


```python
df.isnull().sum()
```

```python
agemean=df["Age"].mean().round(2)
df["Age"].fillna(agemean,inplace=True)
agemean
```

```python
df.info()
```

```python
embarked_mode=df["Embarked"].mode()[0]
df["Embarked"].fillna(embarked_mode)   
df.info()
```

```python
df.isnull().sum()
```

```python
removing_columns=["PassengerId","Name","Ticket"]
```

```python
df=df  .drop(removing_columns,axis=1)
```

```python
df.head()
```

```python
df["FamilySize"]=df["SibSp"]+df["Parch"]+1
```

```python
df=df.drop(columns=["SibSp","Parch"])
```

```python
df.head()
```

```python
#df=df.drop(columns=["Cabin"])
```

```python
df.head()
```

```python
    ### Convert categorical variables to numerical
```

```python
df["Gender"]=df["Sex"].map({"male":0,"female":1})
df=df.drop(columns=["Sex"])
```

```python
df.head()
```

```python
df1=df.copy()
```

```python
df1.head()
```

```python
df1=pd.get_dummies(df1,columns=["Embarked"],dtype=int,drop_first=True)
```

```python
df1.head()
```

```python
def fare_category(fare):
    if fare<15:
        return "Low"
    elif fare<40:
        return "Medium"
    else:
        return "High"
```

```python
df1["Fare_category"]=df["Fare"].apply(fare_category)
```

```python
df1.head()
```

```python
df1["Has_Cabin"]= py.where(df1["Cabin"].notna(),"Yes","No")
df1["Has_Cabin"]=df1["Has_Cabin"].map({"Yes":1,"No":0})
df1["Fare_category"]=df1["Fare_category"].map({"High":1,"Low":0})
```

```python
df1.drop(columns=["Cabin"],inplace=True)
df1.head()
```

```python

cleaned_df=df1.copy()
cleaned_df.head()
```

```python
cleaned_df.to_csv("cleaned_titanic_data.csv",index=False)
```
