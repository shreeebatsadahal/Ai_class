USING MEAN IN NUMPY ARRAY

```python
arr1=np.array([1,5,7,8])
mean=np.mean(arr1)
mean
```

```python
arr2=np.array([[6,7,8,10],[8,3,5,12],[10,23,4,6],[10,12,16,18]])
mean1=np.mean(arr2,axis=0)  #column wise mean
mean2=np.mean(arr2,axis=1)  #row wise mean
print(mean1)
print(mean2)
```
