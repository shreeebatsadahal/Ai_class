```python
import numpy as np
```

Concatenation of array


```python
arr1=np.array([[1,2,3,4]])
arr2=np.array([[3,3,3,3]])  
arr3=np.concatenate((arr1,arr2)) ##axis1,column wise axis0,row wise
print(arr3)
```

```python
arr4=np.concatenate((arr1,arr2),axis=1)
print(arr4)
```

```python
arr5=np.array([[4,5,6,8],[9,9,9,9]])
arr6=np.array([[6,6,6,6],[1,1,1,1]])
np.concatenate((arr5,arr6),axis=1)
np.concatenate((arr5,arr6),axis=0)
```

Vstack() hstack()

```python
np.vstack((arr1,arr2))
```

```python
np.hstack((arr1,arr2))
```
